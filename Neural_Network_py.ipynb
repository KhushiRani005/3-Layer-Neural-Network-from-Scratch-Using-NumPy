```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("/student_data.csv")

print(data.head())
```

```python
X = data[['Hours_Studied','Hours_Slept']].values
```

```python
y = data[['Pass']].values
```

```python
print(X.shape)
print(y.shape)
```

```python
X = X/10

print(X)
```

3-layer neural network.

Input Layer

2 neurons

↓

Hidden Layer 1

5 neurons

↓

Hidden Layer 2

4 neurons

↓

Output Layer

1 neuron

Initialize Weights


```python
np.random.seed(42)

W1 = np.random.randn(2,5)
b1 = np.zeros((1,5))

W2 = np.random.randn(5,4)
b2 = np.zeros((1,4))

W3 = np.random.randn(4,1)
b3 = np.zeros((1,1))
```

W1: Input → Hidden Layer 1
W2: Hidden Layer 1 → Hidden Layer 2
W3: Hidden Layer 2 → Output

Activation Function

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

Forward Propagation

```python
hidden1 = sigmoid(np.dot(X,W1)+b1)

hidden2 = sigmoid(np.dot(hidden1,W2)+b2)

output = sigmoid(np.dot(hidden2,W3)+b3)

print(output)
```

Loss Function

```python
loss = np.mean((y-output)**2)

print("Loss =",loss)
```

Derivative Function

```python
def derivative(x):
    return x*(1-x)
```

Learning Rate


```python
lr = 0.1
```

Train the Network

```python
epochs = 5000

for i in range(epochs):

    # Forward
    hidden1 = sigmoid(np.dot(X,W1)+b1)

    hidden2 = sigmoid(np.dot(hidden1,W2)+b2)

    output = sigmoid(np.dot(hidden2,W3)+b3)

    # Loss
    loss = np.mean((y-output)**2)

    # Error
    error = y-output

    d_output = error*derivative(output)

    error2 = d_output.dot(W3.T)

    d_hidden2 = error2*derivative(hidden2)

    error1 = d_hidden2.dot(W2.T)

    d_hidden1 = error1*derivative(hidden1)

    # Update
    W3 += hidden2.T.dot(d_output)*lr

    W2 += hidden1.T.dot(d_hidden2)*lr

    W1 += X.T.dot(d_hidden1)*lr

    if i%500==0:
        print("Epoch:",i," Loss:",loss)
```

Check Accuracy

```python
prediction = (output>0.5).astype(int)

print(prediction)
```

```python
accuracy = np.mean(prediction==y)

print("Accuracy =",accuracy*100,"%")
```

Predict New Student

```python
new = np.array([[6,8]])

new = new/10
```

```python
h1 = sigmoid(np.dot(new,W1)+b1)

h2 = sigmoid(np.dot(h1,W2)+b2)

result = sigmoid(np.dot(h2,W3)+b3)

print(result)
```

```python
if result>0.5:
    print("Student Will Pass")
else:
    print("Student Will Fail")
```
